```python
import os
import random
import cv2
import numpy as np
from tqdm import tqdm
from PIL import Image
import torchvision.transforms as T
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# pre-processing

IMGS = "C:\\Users\\Jamie\\Documents\\University-Stuffs\\Research\\ATR_RemoveBG\\JPEGImages"
MASKS= "C:\\Users\\Jamie\\Documents\\University-Stuffs\\Research\\ATR_RemoveBG\\RemoveBG"
REBUILD_DATA = True


class DataBuilder:
    def __init__(self, img_size, img_dir, mask_dir):
        self.img_size = img_size
        self.imgs = img_dir
        self.masks = mask_dir
        self.count = 0
        
    def build_training_data(self):
        training_data = []
        self.count = 0
        s = self.img_size

        # file names
        img_list = os.listdir(self.imgs)
        mask_list = os.listdir(self.masks)
        
        # augmentation transforms
        img_aug = T.Compose([
                        T.ToTensor(),
                        T.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0),
                        T.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)), # normalized to [-1, 1]
                        T.RandomErasing(p=0.5, scale=(0.02, 0.2), ratio=(0.3, 3.3), value=0),
                        ])
        mask_aug = T.Compose([T.ToTensor(),])

        for mask in tqdm(mask_list):
            try:
                img = mask.replace("-removebg-preview.png", ".jpg")
                if img in img_list:
                    # fetch RGB images
                    img_path = os.path.join(self.imgs, img)
                    img = Image.open(img_path).convert("RGB")
                    h, w = img.size


                    # fetch binary masks
                    mask_path = os.path.join(self.masks, mask)
                    mask = Image.open(mask_path).split()[-1] # alpha channel
                    mask = mask.resize((h, w), Image.BILINEAR)                    


                    # resizing
                    new_size = (s, s * w // h) if h > w else (s * h // w, s)
                    img = img.resize(new_size, Image.BILINEAR)
                    mask = mask.resize(new_size, Image.BILINEAR)


                    # augmentation & padding
                    img = img_aug(img)
                    mask = mask_aug(mask)
                    dh, dw   = s - new_size[0], s - new_size[1]
                    padding = (dh // 2, dh - (dh // 2), dw // 2, dw - (dw // 2))
                    pad = nn.ConstantPad2d(padding, 0)
                    img = pad(img)
                    mask = pad(mask)
                    if random.random() < 0.5:
                        img = T.functional.hflip(img)
                        mask = T.functional.hflip(mask)


                    data = [img, mask]
                    training_data.append(data)                
                    self.count += 1
#                     if self.count == 100:
#                         break

            except Exception as e:
                pass

        torch.save(training_data, 'data.pt')
        print("Built", self.count, "data points!")



if REBUILD_DATA:
    data_builder = DataBuilder(256, IMGS, MASKS)
    data_builder.build_training_data()
    

```
