```python
import os
import cv2
import numpy as np
from tqdm import tqdm

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# pre-processing

DIR_NAME = "C:\\Users\\Jamie\\Documents\\University-Stuffs\\Research\\pytorch tutorial"
REBUILD_DATA = False


class DogsVCats():
    IMG_SIZE = 50
    CATS = os.path.join(DIR_NAME, "PetImages", "Cat")
    DOGS = os.path.join(DIR_NAME, "PetImages", "Dog")
    TEST = os.path.join(DIR_NAME, "PetImages", "Test")
    LABELS = {CATS: 0, DOGS: 1}
    training_data = []
    catCount = 0
    dogCount = 0
    
    def makeTrainingData(self):
        for label in self.LABELS:
            # label is the path of Cats/Dogs
            print("fetching from...", label)
            for name in tqdm(os.listdir(label)):
                try:
                    # appending
                    path = os.path.join(label, name)
                    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
                    img = cv2.resize(img, (self.IMG_SIZE, self.IMG_SIZE))
                    idx = self.LABELS[label]
                    data = [np.array(img), np.eye(2)[idx]]
                    self.training_data.append(data)
                    
                    # counting
                    if label == self.CATS:
                        self.catCount += 1
                    elif label == self.DOGS:
                        self.dogCount += 1
                        
                except Exception as e:
                    pass
                    
        np.random.shuffle(self.training_data)
        np.save(os.path.join(DIR_NAME, "training_data"), self.training_data)
                    
        print("Cats:", self.catCount)
        print("Dogs:", self.dogCount)




if REBUILD_DATA:
    dogsvcats = DogsVCats()
    dogsvcats.makeTrainingData()
            
```

```python
training_data = np.load(os.path.join(DIR_NAME, "training_data.npy"), allow_pickle=True)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ConvNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 32, 5)
        self.conv2 = nn.Conv2d(32, 64, 5)
        self.conv3 = nn.Conv2d(64, 128, 5)
        self.fc1 = nn.Linear(512, 512)
        self.fc2 = nn.Linear(512, 2)
        
    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv2(x))
        x = F.max_pool2d(x, 2, 2)
        x = F.relu(self.conv3(x))
        x = F.max_pool2d(x, 2, 2)
        x = x.flatten(start_dim=1)
        x = F.relu(self.fc1(x))
        x = F.softmax(self.fc2(x), dim=1)

        #print(x.shape)
        return x
```

```python
conv = ConvNet()
print(conv)
x = torch.rand((50, 50))
x = x.view(-1, 1, 50, 50)
y = conv(x)
y
```

```python
# training
import torch.optim as optim

optimizer = optim.Adam(conv.parameters(), lr=1e-3)
lossFunction = nn.MSELoss()
```

```python
X = torch.Tensor([i[0] for i in training_data]) # .view(-1,50,50)??
X = X/255.
X
```

```python
y = torch.Tensor([i[1] for i in training_data])
y
```

```python
valPct = 0.1
valSize = int(len(X) * valPct)
valSize
```

```python
X_val = X[:valSize]
X_train = X[valSize:]

y_val = y[:valSize]
y_train = y[valSize:]
```

```python
y_val.size()
```

```python

BATCH_SIZE = 200
EPOCHS = 1

for epoch in range(EPOCHS):
    for i in tqdm( range(0, len(X_train), BATCH_SIZE) ):
        batch_X = X_train[i:i+BATCH_SIZE].view(-1, 1, 50, 50)
        batch_y = y_train[i:i+BATCH_SIZE]
        
        conv.zero_grad()
        outputs = conv(batch_X)
        loss = lossFunction(outputs, batch_y)
        loss.backward()
        optimizer.step()
        print(loss)
print(loss)


```

```python
# validation

correct = 0
total = 0

with torch.no_grad():
    for i, img in tqdm(enumerate(X_val)):
        predict = np.argmax(conv(img.view(1,1,50,50)))
        truth = y_val[i][1]
        if predict == truth:
            correct += 1
        total += 1
        
print(correct/total)
```
